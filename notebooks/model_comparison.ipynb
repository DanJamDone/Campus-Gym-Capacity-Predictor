```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# results are stored in both models
models = ['Logistic Regression', 'Random Forest']

# results from each model
accuracy = [73.13, 76.12]
mae = [0.358, 0.358]
rmse = [0.733, 0.773]
f1 = [0.69, 0.70]
precision = [0.69, 0.75]
recall = [0.74, 0.76]
```

```python
# X positions for the bars
x = np.arange(len(models))
width = 0.4

# 6 subplots - one per metric
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

# Flatten so we can loop
axes = axes.flatten()

# All metrics and their labels
metrics = [accuracy, mae, rmse, f1, precision, recall]
titles = [
    'Accuracy % (higher is better)',
    'MAE (lower is better)',
    'RMSE (lower is better)',
    'F1 Macro Avg (higher is better)',
    'Precision (higher is better)',
    'Recall (higher is better)'
]

# Loop through and plot each metric
for i, (metric, title) in enumerate(zip(metrics, titles)):
    axes[i].bar(x, metric, width=width, color=['steelblue', 'darkorange'])
    axes[i].set_title(title)
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(models, rotation=15, ha='right')

# Overall title
fig.suptitle('Model Comparison - Logistic Regression vs Random Forest', fontsize=14)

plt.tight_layout()
plt.show()
```
