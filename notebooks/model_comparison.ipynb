```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# results are stored in both models
models = ['Logistic Regression', 'Random Forest']

# results from each model
accuracy = [73.13, 76.12]
mae = [0.358, 0.358]
rmse = [0.733, 0.773]
```

```python
# create a figure with 3 charts side by side
fig, axes = plt.subplots(1, 3, figsize=(14, 6))

# x positions for the bars
x = np.arange(len(models))

# width of the bar
width = 0.4

# Accuracy chart
axes[0].bar(x, accuracy, width=width, color=['steelblue', 'darkorange'])
axes[0].set_title('Accuracy (%)')
axes[0].set_xticks(x)
axes[0].set_xticklabels(models, rotation=15, ha='right')
axes[0].set_ylabel('Accuracy %')
axes[0].set_ylim(0, 100)

# MAE chart
axes[1].bar(x, mae, width=width, color=['steelblue', 'darkorange'])
axes[1].set_title('MAE (lower is better)')
axes[1].set_xticks(x)
axes[1].set_xticklabels(models, rotation=15, ha='right')
axes[1].set_ylabel('MAE')

# RMSE chart 
axes[2].bar(x, rmse, width=width, color=['steelblue', 'darkorange'])
axes[2].set_title('RMSE (lower is better)')
axes[2].set_xticks(x)
axes[2].set_xticklabels(models, rotation=15, ha='right')
axes[2].set_ylabel('RMSE')

# overall title
fig.suptitle('Model Comparison - Logistic Regression vs Random Forest', fontsize=14)

# adjust the spacing
plt.tight_layout()

# display the plot
plt.show()
```
